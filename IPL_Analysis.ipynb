# Data Cleaning

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop(['method'],axis=1,inplace=True)
```

```python
df.drop(['city'],axis=1,inplace=True)
```

```python
df=df.dropna(subset=['winner'])
```

```python
df['player_of_match'] = df['player_of_match'].fillna('Unknown')
```

```python
df['result_margin']=df['result_margin'].fillna(df['result_margin'].median())
```

```python
df.info()
```

```python
# Cleaning the deliveries dataset
df2['extras_type']= df2['extras_type'].fillna('None')
df2['player_dismissed']= df2['player_dismissed'].fillna('None')
df2['dismissal_kind']= df2['dismissal_kind'].fillna('None')
df2['fielder']= df2['fielder'].fillna('None')
```

```python
df2.info()
```

```python
# Extract year, month, day from the date column.
df['date']= pd.to_datetime(df['date'])
df['year']= df['date'].dt.year
df['month']= df['date'].dt.month
df['day']= df['date'].dt.day

# Drop original date column
df.drop('date',axis=1,inplace=True)
```

```python
# Renaming 'id' to 'match_id' in matches data.
df.rename(columns={'id': 'match_id'}, inplace=True)
```

```python
# Merging the datasets on 'match_id'
merged_data= pd.merge(df2, df, how='inner', on='match_id')
```

```python
merged_data
```

```python
merged_data.info()
```

```python
merged_data.head()
```

#  Analysis and Visualization


```python
# Calculate the number of matches each team played
team_matches= merged_data.groupby('team1')['match_id'].nunique() + merged_data.groupby('team2')['match_id'].nunique()
team_matches
```

```python
# Calculate the number of matches each team has won
team_wins= merged_data.groupby('winner')['match_id'].nunique()
team_wins
```

```python
# Calculate the winning percentage
win_percentage = (team_wins / team_matches) * 100
win_percentage.round(2)
```

```python
# ploting winning percentage
sorted_win_percentage = win_percentage.sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=sorted_win_percentage.index, y=sorted_win_percentage.values, color="blue")
plt.title('Winning Percentage by Team')
plt.xlabel('Team')
plt.ylabel('Winning Percentage')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', color = 'darkred')
plt.show()
```

```python
# Calculate total runs scored by each batsman
batsman_runs= merged_data.groupby('batter')['batsman_runs'].sum()
batsman_runs
```

```python
# top 10 batsmen
top_batsmen= batsman_runs.sort_values(ascending=False).head(10)
top_batsmen
```

```python
# Ploting top 10 batsmen
top_batsmen.plot(kind='bar', color='darkblue')
plt.title('Top 10 Batsmen by Total Runs')
plt.xlabel('Batsmen')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', color = "darkred")
plt.show()
```

```python
wickets= merged_data[merged_data['is_wicket'] == 1]
wickets
```

```python
# Calculate total wickets taken by each bowler
bowler_wickets= wickets.groupby('bowler')['is_wicket'].sum()
bowler_wickets
```

```python
# Get the top 10 bowlers
top_bowlers= bowler_wickets.sort_values(ascending=False).head(10)
top_bowlers
```

```python
# Plot the top 10 bowlers
top_bowlers.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Bowlers by Total Wickets')
plt.xlabel('Bowlers')
plt.ylabel('Total Wickets')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
